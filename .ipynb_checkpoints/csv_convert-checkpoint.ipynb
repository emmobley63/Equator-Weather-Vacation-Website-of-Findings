```python
import pandas as pd
import time
from datetime import datetime


file_path = "Resources/cities.csv"
```

```python
cities_df = pd.read_csv(file_path)

cities_df = pd.DataFrame(cities_df)

cities_df.head(40)
```

```python
city_name_title_case = []

for x in cities_df["City"]:
    city = str(x)
    city = city.title()
    city_name_title_case.append(city)
    
cities_df["City"] = city_name_title_case
cities_df.head()
```

```python
date_time = []

for x in cities_df["Date"]:
    date = datetime.fromtimestamp(x)
    date_time.append(date)

cities_df["Date"] = date_time

cities_df.head()
```
